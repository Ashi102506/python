```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def show_info(self):
        print(f"Name: {self.name}, Marks: {self.marks}")

s1 = Student("Ashi", 100)
s1.show_info()
```

```python
def fact(n):
    if n==0 or n==1:
        return 1
    else:
        return n*fact(n-1)
n=int(input("enter no."))
print(fact(n))
```

```python
def fibonnaci(n)
   if n<=1:
       return n
   elif n==2:
       return 1
   else :
       return fibonnaci(n-1)+fibonnaci(n-2)
num=int(input("enter no. of terms in fibonnaci"))
for i in range(num):
    print(fibonnaci(i))
```
